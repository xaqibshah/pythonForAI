```python
#load dataset of titanic seaborn
df = sns.load_dataset('titanic')
 
# df = pd.load_dataset('titanic')
```

```python
df.head()
```

```python
sns.histplot(df['parch'], kde=False)
plt.show()
```

```python
sns.distplot(df['age'], hist=False)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.distplot(df['age'])
plt.show()
```

```python
sns.relplot(x='age', y='fare', col='pclass', hue='sex', style='sex', kind='line', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='age', y='fare',hue='sex', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.lineplot(x='age', y='fare', hue='sex',style='sex', data=df)
plt.show()
```

```python
# plt.figure(figsize=(8,8))
sns.barplot(x='sex', y='survived', hue='pclass', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.stripplot(x='sex', y='age', hue='sex',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.swarmplot(x='sex', y='age', hue='sex', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x='survived', y='age', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x='age', y='survived', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.violinplot(x='survived', y='age',hue='sex', data=df)
plt.show()
```

```python
sns.countplot(x='survived', data=df, palette='Blues')
```

```python
plt.subplots(figsize=(8,8))
sns.pointplot(x='sex', y='survived', hue='pclass', data=df)
plt.show()
```

```python
sns.lmplot(x='age', y='fare', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.regplot(x='fare', y='fare', data=df)
plt.show()
```

```python
# Convert categorical columns to numeric values
df['sex'] = df['sex'].map({'female': 0, 'male': 1})

plt.figure(figsize=(8,8))
sns.heatmap(df.corr(), annot=True)
plt.show()
```
